```python
from tkinter import *
from tkinter import ttk
from tkinter import messagebox
from functools import partial
import numpy as np
import time
import pandas as pd
import smtplib

data=pd.read_csv('converted_training_data.csv')
test_data=pd.read_csv('converted_test_data.csv')
data=data.drop('Loan_Status',axis=1)

df=pd.concat([data,test_data])
df.LoanAmount=df.LoanAmount.apply(lambda x:x+10000)
X=df.drop(['LoanAmount','Property_Area','Education'],axis=1)
y=df.LoanAmount
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(X,y)


win=Tk()
win.title("Loan Amount Processing Window")
win.config(bg="brown2")
root=Frame(win,bg="brown2",bd=15)
root.pack(side=RIGHT,fill=Y)

baba="black"
root=root
clr="brown2"

le_d={'MALE':1,'FEMALE':0,'OTHERS':2,'YES':1,'NO':0,'0':0,'1':1,
     '2':2,'3':3,'3+':3,'SELF':2,'ORGANIZATION':0,'UNEMPLOYED':1}

def data_pre(n,g,m,d,s,ai,ei,la,lat,llp):
    gen=le_d[g]
    mar=le_d[m]
    dep=le_d[d]
    sel=le_d[s]
    last=le_d[llp]
    return n,la,np.array([gen,mar,dep,sel,ai,ei,lat,last]).reshape(1,-1)


def get_data():
    name= e0.get().upper()
    mail=e1.get()
    gen=e2.get()
    mar=e3.get()
    dep=e10.get()
    emp=e4.get()
    apin=int(e5.get())
    exin=int(e6.get())
    loan_amt=int(e7.get())
    loan_amt_t=int(e8.get())
    l_l_p=e9.get()
    
    
    usr_data=data_pre(name,gen,mar,dep,emp,apin,exin,loan_amt,loan_amt_t,l_l_p)
    rec=[mail]
    con=f'''Dear {name},
    We are pleased to inform you that the loan amount which your were requesting
    has been processed.
    Bring the documents and take the Loan of {loan_amt}   BAM!!!'''
    server=smtplib.SMTP('smtp.gmail.com',587)
    server.ehlo()
    server.starttls()
    server.login('sender mail id','password')
    
    usr_data=data_pre(name,gen,mar,dep,emp,apin,exin,loan_amt,loan_amt_t,l_l_p)

    pre=model.predict(usr_data[2])
    if(pre[0]<usr_data[1]):
        l11=Label(root,text=f'Hello {usr_data[0]} you are not elible to get {usr_data[1]} amount loan',width=70,font=("Comic Sans MS",10),bg=clr,fg="black")
        l11.grid()
        print(f'Hello {usr_data[0]} you are not elible to get {usr_data[1]} amount loan')
        time.sleep(5)
        l12=Label(root,text=f'But you can take {int(pre[0])} amount as a loan!!!',width=60,font=("Comic Sans MS",10),bg=clr,fg="black")
        l12.grid()
        print(f'But you can take {int(pre[0])} amount as a loan!!!')
    else:
        l11=Label(root,text=f'Congratulations {usr_data[0]} you can take {usr_data[1]} amount loan!,Details are on your mail',width=70,font=("Comic Sans MS",10),bg=clr,fg="black")
        l11.grid()
        print(f'Congratulations {usr_data[0]} you can take {usr_data[1]} amount loan!,Details are on your mail')
        try:
            server.sendmail('sender mai-id',rec,con)
        except Exception as e:
            print(e)
     
label1=Label(root,text="Enter Your Details",font=("Comic Sans MS",20),bg=clr,fg="black")
label1.grid()
l0=Label(root,text="Full Name",width=30,font=("Comic Sans MS",10),bg=clr,fg="black")
l0.grid()
e0=Entry(root,bg="white",font=20,width=30)
e0.grid()

l1=Label(root,text="Mail Id",width=30,font=("Comic Sans MS",10),bg=clr,fg="black")
l1.grid()
e1=Entry(root,bg="white",font=20,width=30)
e1.grid()
l2=Label(root,text="Gender",width=30,font=("Comic Sans MS",10),bg=clr,fg="black")
l2.grid()
e2=ttk.Combobox(root,state="readonly",font=10,width=30)
e2['value']=('MALE', 'FEMALE', 'OTHERS')
e2.current(0)    
e2.grid()
l3=Label(root,text="Are You Married?",width=30,font=("Comic Sans MS",10),bg=clr,fg="black")
l3.grid()
e3=ttk.Combobox(root,state="readonly",font=20,width=30)
e3['value']=('YES','NO')
e3.current(0)
e3.grid()
l10=Label(root,text="please enter the dependents",width=30,font=("Comic Sans MS",10),bg=clr,fg="black")
l10.grid()
e10=ttk.Combobox(root,state="readonly",font=20,width=30)
e10['value']=('0','1', '2', '3+')
e10.current(0)
e10.grid()
l4=Label(root,text="Employment Status",width=30,font=("Comic Sans MS",10),bg=clr,fg="black")
l4.grid()
e4=ttk.Combobox(root,state="readonly",font=10,width=30)
e4['value']=('SELF', 'ORGANIZATION', 'UNEMPLOYED')
e4.current(0)
e4.grid()
l5=Label(root,text="Yearly Income",width=30,font=("Comic Sans MS",10),bg=clr,fg="black")
l5.grid()
e5=Entry(root,bg="white",font=20,width=30)
e5.grid()
l6=Label(root,text="Extra Income",width=30,font=("Comic Sans MS",10),bg=clr,fg="black")
l6.grid()
e6=Entry(root,bg="white",font=20,width=30)
e6.grid()
l7=Label(root,text="Loan Amount",width=30,font=("Comic Sans MS",10),bg=clr,fg="black")
l7.grid()
e7=Entry(root,bg="white",font=20,width=30)
e7.grid()
l8=Label(root,text="Loan Amount Period in Days(Min 12 days)",width=40,font=("Comic Sans MS",10),bg=clr,fg="black")
l8.grid()
e8=Entry(root,bg="white",font=20,width=30)
e8.grid()
l9=Label(root,text="Did you pay your last loan?",width=30,font=("Comic Sans MS",10),bg=clr,fg="black")
l9.grid()
e9=ttk.Combobox(root,state="readonly",font=20,width=30)
e9['value']=('YES','NO')
e9.current(0)
e9.grid()
b1=Button(root,text="Check Amount",bg="slateGray2",width=20,font=("bold",10),
                   command=get_data)
b1.grid()

win.mainloop()
```
